```python
import numpy as np
import pandas as pd
df = pd.read_csv('./UCI_Credit_Card.csv')
df = df.rename(columns={'default.payment.next.month': 'default', 'PAY_0': 'PAY_1'})
df = df.drop('ID',axis=1)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
df.describe()
```

```python
classes = df['default'].unique()

# Create an empty DataFrame to store the sampled data
sampled_df = pd.DataFrame(columns=df.columns)

# # Sample half of each class
for class_label in classes:
    class_data = df[df['default'] == class_label]
    num_samples_to_keep = 1000
    sampled_data = class_data.sample(n=1000, random_state=37)
    sampled_df = pd.concat([sampled_df, sampled_data])

# # Reset the index of the sampled DataFrame
sampled_df.reset_index(drop=True, inplace=True)
```
